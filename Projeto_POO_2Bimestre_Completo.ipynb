```python
class Emilly_viagensplanejadas:
    def __init__(self, destino, data, status):
        self.__destino = destino
        self.data = data
        self.status = status

    def get_destino_Emilly(self):
        return self.__destino

    def set_destino_Emilly(self, novo_destino):
        if isinstance(novo_destino, str) and novo_destino.strip() != "":
            self.__destino = novo_destino

    def exibir_dados_Emilly(self):
        return f"Destino: {self.__destino}, Data: {self.data}, Status: {self.status}"

    def viagem_confirmada_Emilly(self):
        return self.status.lower() == "confirmada"


viagem1 = Emilly_viagensplanejadas("Brasil", "2025-07-10", "Planejada")
viagem1.set_destino_Emilly("Argentina")

print(viagem1.get_destino_Emilly())
print(viagem1.exibir_dados_Emilly())
print(viagem1.viagem_confirmada_Emilly())
```

## PARTE 4 ‚Äì Polimorfismo e Banco de Dados

**Tarefas:**
- Implementar polimorfismo: sobrescrever um m√©todo na subclasse.
- Integrar o sistema com banco de dados (MySQL ou SQLite).
- Implementar inser√ß√£o (INSERT), consulta (SELECT) e exclus√£o (DELETE) de registros no banco de dados.

**C√≥digo:**

```python
import mysql.connector

class Viagem_Emilly:
    def __init__(self, destino, data, status):
        self.destino = destino
        self.data = data
        self.status = status

    def exibir_dados_Emilly(self):
        return f"Destino: {self.destino}, Data: {self.data}, Status: {self.status}"

class ViagemEspecial_Emilly(Viagem_Emilly):
    def __init__(self, destino, data, status, guia):
        super().__init__(destino, data, status)
        self.guia = guia

    def exibir_dados_Emilly(self):
        return f"Destino: {self.destino}, Data: {self.data}, Status: {self.status}, Guia: {self.guia}"

class BancoDados_Emilly:
    def __init__(self, host='localhost', database='emilly_viagensplanejadas', user='root', password=''):
        try:
            self.conn = mysql.connector.connect(
                host=host,
                database=database,
                user=user,
                password=password
            )
            if self.conn.is_connected():
                self.cursor = self.conn.cursor()

                self.cursor.execute('''
                    CREATE TABLE IF NOT EXISTS viagens (
                        id INT AUTO_INCREMENT PRIMARY KEY,
                        destino VARCHAR(255),
                        data DATE,
                        status VARCHAR(50)
                    )
                ''')
                self.conn.commit()
                print("Conectado ao banco e tabela criada/verificada.")
        except Error as e:
            print(f"Erro ao conectar no banco de dados: {e}")

    def inserir_viagem_Emilly(self, viagem):
        sql = "INSERT INTO viagens (destino, data, status) VALUES (%s, %s, %s)"
        val = (viagem.destino, viagem.data, viagem.status)
        self.cursor.execute(sql, val)
        self.conn.commit()

    def consultar_viagens_Emilly(self):
        self.cursor.execute('SELECT * FROM viagens')
        return self.cursor.fetchall()

    def deletar_viagem_Emilly(self, id_viagem):
        sql = "DELETE FROM viagens WHERE id = %s"
        self.cursor.execute(sql, (id_viagem,))
        self.conn.commit()

    def fechar_Emilly(self):
        if self.conn.is_connected():
            self.cursor.close()
            self.conn.close()
            print("Conex√£o com o banco encerrada.")


db = BancoDados_Emilly(
    host='localhost',
    database='emilly_viagensplanejadas',
    user='root',
    password=''  
)
v1 = Viagem_Emilly("Brasil", "2025-07-10", "Planejada")
v2 = ViagemEspecial_Emilly("Argentina", "2025-08-15", "Confirmada", "Guia Emilly")

db.inserir_viagem_Emilly(v1)
db.inserir_viagem_Emilly(v2)

viagens = db.consultar_viagens_Emilly()
for viagem in viagens:
    print(f"ID: {viagem[0]}, Destino: {viagem[1]}, Data: {viagem[2]}, Status: {viagem[3]}")

db.deletar_viagem_Emilly(1) 

db.fechar_Emilly()
```
