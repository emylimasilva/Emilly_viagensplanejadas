```python
class Viagem:
    def __init__(self, destino, data, status):
        self.destino = destino
        self.data = data
        self.status = status

    def alterar_status(self, novo_status):
        self.status = novo_status

    def reagendar(self, nova_data):
        self.data = nova_data

    def exibir_detalhes(self):
        print(f"Destino: {self.destino}")
        print(f"Data: {self.data}")
        print(f"Status: {self.status}")
        print("-" * 30)

class ViagemInternacional(Viagem):
    def __init__(self, destino, data, status):
        super().__init__(destino, data, status)


viagem3 = ViagemInternacional("Canad√°", "2025-09-10", "planejada")
viagem3.exibir_detalhes()
```

```python
import mysql.connector

class Viagem:
    def __init__(self, destino, data, status):
        self.destino = destino
        self.data = data
        self.status = status

    def exibir_detalhes(self):
        print(f"Destino: {self.destino}")
        print(f"Data: {self.data}")
        print(f"Status: {self.status}")
        print("-" * 30)

class BancoDadosEmilly:
    def __init__(self, host='localhost', database='godream', user='root', password=''):
        self.conn = mysql.connector.connect(
            host=host,
            database=database,
            user=user,
            password=password
        )
        self.cursor = self.conn.cursor()

    def inserir_viagem(self, viagem):
        sql = "INSERT INTO viagens (destino, data, status) VALUES (%s, %s, %s)"
        val = (viagem.destino, viagem.data, viagem.status)
        self.cursor.execute(sql, val)
        self.conn.commit()

    def consultar_viagens(self):
        self.cursor.execute("SELECT id, destino, data, status FROM viagens")
        return self.cursor.fetchall()

    def deletar_viagem(self, id_viagem):
        sql = "DELETE FROM viagens WHERE id = %s"
        self.cursor.execute(sql, (id_viagem,))
        self.conn.commit()

    def fechar(self):
        self.cursor.close()
        self.conn.close()

def menu():
    db = BancoDadosEmilly()

    while True:
        print("\n--- Menu Viagens ---")
        print("1. Inserir viagem")
        print("2. Consultar todas as viagens")
        print("3. Deletar viagem por ID")
        print("4. Sair")

        escolha = input("Escolha uma op√ß√£o: ")

        if escolha == '1':
            destino = input("Destino: ")
            data = input("Data (AAAA-MM-DD): ")
            status = input("Status: ")
            viagem = Viagem(destino, data, status)
            db.inserir_viagem(viagem)
            print("Viagem inserida!")

        elif escolha == '2':
            viagens = db.consultar_viagens()
            if viagens:
                for v in viagens:
                    print(f"ID: {v[0]}, Destino: {v[1]}, Data: {v[2]}, Status: {v[3]}")
            else:
                print("Nenhuma viagem cadastrada.")

        elif escolha == '3':
            id_del = input("Digite o ID da viagem para deletar: ")
            db.deletar_viagem(id_del)
            print("Viagem deletada (se o ID existir).")

        elif escolha == '4':
            print("Encerrando...")
            db.fechar()
            break

        else:
            print("Op√ß√£o inv√°lida. Tente novamente.")

menu()
```
